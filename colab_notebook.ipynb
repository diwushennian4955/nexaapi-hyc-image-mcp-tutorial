```python
# Cell 2: Setup NexaAPI client
# Get your FREE API key at: https://rapidapi.com/user/nexaquency
# 100 free images, no credit card required!

from nexaapi import NexaAPI
import requests
from IPython.display import Image, display, Audio
import os

# Set your API key here
NEXAAPI_KEY = 'YOUR_RAPIDAPI_KEY'  # Get free key: https://rapidapi.com/user/nexaquency

# Initialize client
client = NexaAPI(api_key=NEXAAPI_KEY)

print('✅ NexaAPI client initialized!')
print('💰 Pricing: $0.003/image — 10x cheaper than DALL-E 3')
print('🔗 Website: https://nexa-api.com')
```

```python
# Cell 3: Generate an image with NexaAPI
# This costs only $0.003 — 56+ models available!

prompt = 'A business invoice document with company name, date, itemized costs and total amount'
model = 'flux-schnell'  # Options: flux-schnell, flux-pro, stable-diffusion-3

print(f'🎨 Generating image with NexaAPI...')
print(f'📝 Prompt: {prompt}')
print(f'🤖 Model: {model}')
print(f'💰 Cost: $0.003')

response = client.image.generate(
    model=model,
    prompt=prompt,
    width=1024,
    height=1024
)

image_url = response.image_url
print(f'\n✅ Image generated!')
print(f'🔗 URL: {image_url}')
```

```python
# Cell 4: Download and display the generated image

import io
from PIL import Image as PILImage
import matplotlib.pyplot as plt

# Download image
img_response = requests.get(image_url, timeout=30)
img_response.raise_for_status()

# Save locally
with open('generated_image.png', 'wb') as f:
    f.write(img_response.content)

# Display
img = PILImage.open(io.BytesIO(img_response.content))
plt.figure(figsize=(10, 10))
plt.imshow(img)
plt.axis('off')
plt.title(f'Generated by NexaAPI ({model}) — Cost: $0.003', fontsize=14)
plt.tight_layout()
plt.show()

print(f'✅ Image saved as generated_image.png')
print(f'📐 Size: {img.size[0]}x{img.size[1]} pixels')
```
